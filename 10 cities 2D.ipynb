```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
```

```python
cities=["Tehran","Isfahan","Tabriz","Mashhad","Shiraz","Qom","Karaj","Sanandaj","Yazd","Ilam"]
X=[51.2323,51.4013,46.1757,59.3236,52.3233,50.5235,50.5656,46.5946,54.2023,46.2426]
Y=[35.4120,32.3955,38.0454,36.1935,29.3636,34.3824,35.4938,35.1841,31.5256,33.3806]
population=[16800,2200,1700,3400,1995,1200,1970,414,529,194]
colors=["red","blue","black","green","pink","yellow","gray","cyan","purple","brown"]
```

```python
for i,city in enumerate(cities):
    plt.scatter(X[i],Y[i],s=np.array(population[i]),c=colors[i],alpha=0.7)

plt.xlim([45.5,60])
plt.ylim([29,38.5])
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.title("populations of 10 big cities of Iran")
legend_elements=[]
for i,city in enumerate(cities):
    patch=mpatches.Patch(color=colors[i],label=city)
    legend_elements.append(patch)
plt.legend(handles=legend_elements)
plt.show()
```

```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
# create new figure, axes instances.
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
# setup mercator map projection.
m = Basemap(llcrnrlon=47.,llcrnrlat=26.,urcrnrlon=63.,urcrnrlat=40.,\
            rsphere=(6378137.00,6356752.3142),\
            resolution='l',projection='merc',\
            lat_0=40.,lon_0=-20.,lat_ts=20.,ax=ax)

cities=["Tehran","Isfahan","Tabriz","Mashhad","Shiraz","Qom","Karaj","Sanandaj","Yazd","Ilam"]
X=[51.2323,51.4013,46.1757,59.3236,52.3233,50.5235,50.5656,46.5946,54.2023,46.2426]
Y=[35.4120,32.3955,38.0454,36.1935,29.3636,34.3824,35.4938,35.1841,31.5256,33.3806]
population=[16800,2200,1700,3400,1995,1200,1970,414,529,194]
colors=["red","blue","black","green","pink","yellow","gray","cyan","purple","brown"]

x,y=m(X,Y)
#m.scatter(x,y,population,ax=ax)
for i,city in enumerate(cities):
    m.scatter(X[i],Y[i],s=np.array(population[i]),c=colors[i],alpha=0.7)
m.drawcoastlines(ax=ax)
m.fillcontinents(ax=ax)
# draw parallels
m.drawparallels(np.arange(25,50,11),labels=[1,1,0,1],ax=ax)
# draw meridians
m.drawmeridians(np.arange(-180,90,5),labels=[1,1,0,1],ax=ax)
plt.xlabel("longitude")
plt.ylabel("latitude")
ax.set_title("populations of 10 big cities of Iran")
legend_elements=[]
for i,city in enumerate(cities):
    patch=mpatches.Patch(color=colors[i],label=city)
    legend_elements.append(patch)
plt.legend(handles=legend_elements)
plt.show()
```
