```python
import cudf as pd
pd.__version__
```

```python
df = pd.read_csv("/content/US_Accidents_March23.csv")
df.info()
```

```python
df.describe()
```

```python
import numpy as np

df["Start_Time"] = pd.to_datetime(df["Start_Time"].str[:19])

df
# 2016-02-08 11:00:00
# 26 2022-04-29 08:25:22.000000
# 29 2017-07-23 04:21:01.000000000
```

```python
from sklearn.preprocessing import LabelEncoder

city_encoder = LabelEncoder()
df["CityEncode"] = city_encoder.fit_transform(df["City"].to_pandas())

Weather_Condition_Encode = LabelEncoder()
df["Weather_Condition_Encode"] = Weather_Condition_Encode.fit_transform(df["Weather_Condition"].to_pandas())

State_Encode = LabelEncoder()
df["State_Encode"] = State_Encode.fit_transform(df["State"].to_pandas())

Wind_Direction_Encode = LabelEncoder()
df["Wind_Direction_Encode"] = Wind_Direction_Encode.fit_transform(df["Wind_Direction"].to_pandas())
```

```python
df[["City", "CityEncode"]]
df[["Weather_Condition", "Weather_Condition_Encode"]]
df[["State", "State_Encode"]]
df[["Wind_Direction", "Wind_Direction_Encode"]]
```

```python
df["Temperature(F)"].fillna(df["Temperature(F)"].median(), inplace=True)
df["Humidity(%)"].fillna(df["Humidity(%)"].median(), inplace=True)
df["Pressure(in)"].fillna(df["Pressure(in)"].median(), inplace=True)
df["Visibility(mi)"].fillna(df["Visibility(mi)"].median(), inplace=True)
df["Wind_Speed(mph)"].fillna(df["Wind_Speed(mph)"].median(), inplace=True)
df["Wind_Direction_Encode"].fillna(df["Wind_Direction_Encode"].median(), inplace=True)

df.info()
```

```python
columns_to_drop = [
    "ID", "Source", "End_Lat", "End_Lng", "Country", "State", "City", "Wind_Direction", "Weather_Timestamp", "End_Time", "Description", "Street",
    "County", "Zipcode", "Airport_Code", "Sunrise_Sunset", "Civil_Twilight", "Nautical_Twilight", "Astronomical_Twilight", "Wind_Chill(F)", "Timezone", "Weather_Condition"
]

df.drop(columns=columns_to_drop, inplace=True)


print(df.head())
print(df.info())
```

```python
df[df.select_dtypes(include=['int']).columns] = df.select_dtypes(include =['int']).astype("uint8")
df[df.select_dtypes(include=['float']).columns] = df.select_dtypes(include =['float']).astype("float32")
df[df.select_dtypes(include=['bool']).columns] = df.select_dtypes(include =['bool']).astype("uint8")

print(df.head())
print(df.info())
```
